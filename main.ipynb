```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('superstore_sales.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df['Price'].fillna(df['Price'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Profit'].fillna(df['Profit'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
plt.Figure(figsize=(10,6))
sns.boxplot(x=df['Price'],y=df['Profit'])
plt.title('outliers in price and Profit')
plt.show()
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3- Q1

df = df[(df['Price'] >= (Q1 - 1.5*IQR)) & (df['Price']<= (Q3 + 1.5*IQR))]
df
df
```

```python
df['Revenue'] = df['Price']*df['Quantity']
```

```python
df.head(20)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

```

```python
df.head(20)
```

```python
if 'Category' in df.columns:
    df = pd.get_dummies(df, columns=['Category'])
else:
    print("Column 'Category' not found in the dataset")
```

```python
df.head(20)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df_numeric = df.select_dtypes(include=['number'])

plt.Figure(figsize= (10,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap= 'coolwarm',)
plt.title('correlation analysis')
plt.show()
```

```python
df.columns
```

```python
df.groupby('Month')['Revenue'].sum().plot(kind='bar', figsize=(10,6), title="Monthly Revenue")
```
