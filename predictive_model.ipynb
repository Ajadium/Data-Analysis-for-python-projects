```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv('house_prices.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.pairplot(df, diag_kind= "kde")
plt.show()
```

```python
x = df[['Square_Feet', 'Bedrooms', 'Age']] # define feature x
y = df['Price'] # target variable y
```

```python
X_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42) #split data into train and test
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error (y_test, y_pred)
```

```python
print(f'(mean_absolute_errro): {mae}')
print(f'(mean_squre_errro): {mse}')
```

```python
plt.scatter(y_test, y_pred, alpha= 0.5)
plt.ylabel('actual value')
plt.xlabel('predicted value')
plt.title('actual value vs predicted value')
plt.show()
```
